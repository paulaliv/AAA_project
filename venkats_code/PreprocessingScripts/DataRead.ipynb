```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

os.chdir('/home/vsayyalasomayajula/Documents/Proteomics/PlasmaAAA/')

raw_data_file = 'proteomics data, removed poor QC samples.xlsx'

RawProteinData = pd.read_excel(raw_data_file, sheet_name='Sheet1')
PatientData = RawProteinData.iloc[:, 1:]

print(RawProteinData.head(10))
print(PatientData.head(10))
print(PatientData.shape[0])
```

```python
missing_data_summary = PatientData.isnull().sum()
print(missing_data_summary[missing_data_summary > 0])
```

```python
# Calculate percentage of missing data
missing_percentage = (PatientData.isnull().sum() / len(PatientData)) * 100
print(missing_percentage[missing_percentage > 0].sort_values(ascending=False))
```

```python
string_hgnc = pd.read_csv("/home/vsayyalasomayajula/Documents/Proteomics/PPI_STRING/ensembl_to_hgnc.csv")
string_proteins = set(string_hgnc["HGNC_Symbol"].dropna().unique())
Analyzed_proteins = set(RawProteinData['Genes'])
#
common_proteins = Analyzed_proteins.intersection(string_proteins)
missing_proteins = Analyzed_proteins - string_proteins
#
print(f"Analyzed proteins: {len(Analyzed_proteins)}")
print(f"Overlap with STRING: {len(common_proteins)}")
print(f"Missing in STRING: {len(missing_proteins)}")
```

```python
print("Sample of missing proteins:", missing_proteins) 
```

Remove proteins that are not identified in STRING or UNIPROT databases

```python
RawProteinData = RawProteinData[~RawProteinData['Genes'].isin(missing_proteins)]
PatientData = RawProteinData.iloc[:, 1:]
print(RawProteinData.head(10))
print(PatientData.head(10))
print(PatientData.shape[0])
```
